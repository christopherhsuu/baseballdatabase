```python
import pandas as pd
import seaborn as sns
import matplotlib as plt
```

```python
hit = pd.read_csv("2023hitterstats.csv")
```

```python
#remove pitchers from the list

def remove_pitchers(hit):
    filtered_hit = hit[hit['Pos Summary'] != '1']
    filtered_hit.reset_index(drop=True,inplace=True)
    filtered_hit['Rk'] = filtered_hit.index +1
    return filtered_hit

hit = remove_pitchers(hit)
```
